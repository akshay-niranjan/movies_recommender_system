```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# genres
# id
# keyword
# title
# overview
# budget,release_date,revenue - not taking numerical columns
# cast
# crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
# Data preprocessing 
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
# correcting format of cols
# string -> list of dictionary -> array
```

```python
import ast

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):     # string -> list of dictionary
        L.append(i['name'])
    return L
```

```python
movies.iloc[0].genres
```

```python
movies.genres = movies.genres.apply(convert)
movies.keywords = movies.keywords.apply(convert)
```

```python
movies.sample(2)
```

```python
movies.crew[0]     # we only need director from "job":crew
```

```python
# Taking top 3 cast & only real name

def convert_top3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):  # string -> list of dictionary
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L

movies.cast = movies.cast.apply(convert_top3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):  # string -> list of dictionary
        if i['job']=='Director':
            L.append(i['name'])
        else:
            break
    return L

movies.crew = movies.crew.apply(fetch_director)
```

```python
movies.sample(2)
```

```python
movies.crew 
```

```python
movies.overview[0]
```

```python
# convert overview into list
movies.overview = movies.overview.apply(lambda x:x.split())
```

```python
# Removing space from names 
movies.genres = movies.genres.apply(lambda x:[i.replace(" ","") for i in x])
movies.keywords = movies.keywords.apply(lambda x:[i.replace(" ","") for i in x])
movies.crew = movies.crew.apply(lambda x:[i.replace(" ","") for i in x])
movies.cast = movies.cast.apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies.overview+movies.genres+movies.keywords+movies.cast+movies.crew
df = movies[['movie_id','title','tags']]
```

```python
df
```

```python
#convert tags list to string
df.tags=df.tags.apply(lambda x:" ".join(x))
```

```python
df.tags=df.tags.apply(lambda x:x.lower())
```

```python
df.tags[0]
```

```python
df.head()
```

```python
# Removing same words like ability abilities instead taking only one
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))         #converting words into root words

    return " ".join(y)

df.tags=df.tags.apply(stem)
```

```python
# Text Vectorization -> bag of words -> combine all tags -> 5000 most common words
# stop words(are and of is) are not considered -> they should be removed
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df.tags).toarray()     # sparse matrix -> to array
```

```python
vectors
```

```python
list(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity     # calculating cosine distances
similarity = cosine_similarity(vectors)              # 4806 movies distance with 4806 movies
```

```python
similarity[0]
```

```python
# recommend function 
# find index
# find similarity[index]
# sort distances in similarity
# fetch top 5 & recommend

def recommend(movie):
    movie_index = df[df.title == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(df.iloc[i[0]].title)

    return
```

```python
recommend('Interstellar')
```
